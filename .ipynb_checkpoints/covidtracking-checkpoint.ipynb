```python
# to handle  data retrieval
import requests

# to handle certificate verification
import certifi

# to manage json data
import json

# for pandas dataframes
import pandas as pd
```

```python
# URL for GET requests to retrieve vehicle data
url = "https://covidtracking.com/api/states/daily"

# Perform a get request for this character
response = requests.get(url)
```

```python
# Storing the JSON response within a variable
data = response.json()
# Use json.dumps to print the json
# print(json.dumps(data, indent=4, sort_keys=True))
```

```python
# in this dataset, the data to extract is under 'features'
covidtracking_df = pd.json_normalize(data)
covidtracking_df
```

```python
covidtracking_df.to_csv (r'covidtracking_dataframe.csv', index = False, header=True)
```
